### visualize the distributions of ratings and compensation

following plots:
  1. scatter plot of employee age and salary, also of healthy_eating and active_lifestyle
  2. countplot of groups they belong to
  3. histogram of employee salary

```python
#1.1

plt.figure(figsize=(10,6))

plt.scatter(data=data, x='age',y='salary', color='b',
           edgecolor='pink', alpha=0.5)
plt.xlabel('Employee Age')
plt.ylabel('Salary of the Employee')
plt.show()
```

#### Observation:
As the age increases, salary also increases, ignoring the outliers.
(the darker dots are due to the fact that some of the values are overlapping for two or more individuals.

```python
#1.2

plt.figure(figsize=(10,6))

plt.scatter(data=data, x='healthy_eating',y='active_lifestyle', color='g',
           edgecolor='linen', alpha=0.5)
plt.xlabel('Healthy Eating')

plt.ylabel('Active Lifestyle')
plt.show()
```

 ^Similar Observation

```python
#1.3

plt.figure(figsize=(10,6))

plt.scatter(data=data, x='healthy_eating',y='salary', color='g',
           edgecolor='linen', alpha=0.5)
plt.xlabel('Healthy Eating')

plt.ylabel('Salary')
plt.show()
```

#### observing that employees w healthy eating habits have higher salary and vice-versa, that is, employees w higher salary try to maintain healthy eating habits.

```python
#2
#Categorical Data
plt.figure(figsize=(10,6))
sns.countplot(data=data, x='groups',
              edgecolor='linen', alpha=0.5)
plt.xlabel('BloodGroups of Employees', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()
```

#### Observations:
Most Employees are of groups 'A' and 'O'.
There are equal number of Employees having groups 'AB' and 'B'.

```python
#3
#Histograms have bins
#Salary distribution

plt.figure(figsize=(10,6))
sns.histplot(data=data, x='salary', color='brown',
            edgecolor='linen', alpha=0.9, bins=10)
plt.xlabel('Salary Distribution')
plt.ylabel('Count of Employees for that Salary range')
plt.show()
```

#### Observation:
 1. number of employee having higher salary are low. 
 2. The distribution is not uniform, but random
 3. Large number of employees are under salary range of 1000 to 3000
